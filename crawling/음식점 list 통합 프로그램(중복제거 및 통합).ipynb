```python
import os
import pandas as pd

first_flag = 1
os.chdir(r'C:\Users\AI\Desktop\훗카이도')
location = os.listdir()
for i in location:
    print(i)
    os.chdir(i)
    os.chdir('list')
    files = os.listdir()
    if len(files) == 0:
        os.chdir('..')
        os.chdir('..')
        continue
    for j in files:
        sub = pd.read_csv(j,names=['title','count','url','scrap_flag'])
        if first_flag == 1:
            main = pd.read_csv(j,names=['title','count','url','scrap_flag'])
            first_flag = 0
        else:
            main = pd.concat([main,sub])
    main.drop_duplicates('url')
    main.to_csv('restaurant_urls_{} 음식점.csv'.format(i),index=False,encoding='utf-8')
    first_flag = 1
    del main
    os.chdir('..')
    os.chdir('..')
```
