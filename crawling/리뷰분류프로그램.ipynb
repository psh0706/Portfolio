```python
import pandas as pd

for i in range(0,38):
    file = pd.read_csv('restaurant_review_{}.csv'.format(i))
    english = file[file['lang']=='en']
    korean = file[file['lang']=='ko']
    english.to_csv('restaurant_review_{}_en.csv'.format(i),index=False,encoding='utf-8')
    korean.to_csv('restaurant_review_{}_ko.csv'.format(i),index=False,encoding='utf-8')
```
