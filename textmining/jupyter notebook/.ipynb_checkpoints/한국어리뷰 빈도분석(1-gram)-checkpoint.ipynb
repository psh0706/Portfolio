```python
from collections import Counter
from konlpy.tag import Hannanum
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = []
review = []
Noun = []
a = 1

import re

emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)



twitter= Hannanum()

data = pd.read_csv(r"C:\Users\leejisan\Desktop\machine_learning\sample_review\food\SEEN Restaurant & Bar Bangkok.csv",encoding='utf-8',engine='python')
for i in range(len(data)):
    try:
        data['content'][i] = emoji_pattern.sub('', data['content'][i])
    except:
        print(data['content'][i])
        print(i)
    
text = data['content']

stop_words = ['&','것','구','3','2','그',"It's","I'm",'수','5','우리',
              '고','듯','구','거','차','의','은','4','편','10','분',
             '개','번','저','여기','다.','30','나',"I've",'저희때','라','후',
              '가능','곳','8','만','A5','두','몇','을','한','네','오','1','20','중'
             ,'#1','를','+','aS','25','동안','이곳','시도','때문','이것','일부','때','그것','거리']


for line in text:
    morph = twitter.pos(line)
    review.append(morph)

for i in review:
    for word , tag in i:
        if tag in 'Noun':
            if not i in stop_words:
                Noun.append(word)
            
Noun = [i for i in list((Noun)) if not i in stop_words]  # 불영어 제거
```
