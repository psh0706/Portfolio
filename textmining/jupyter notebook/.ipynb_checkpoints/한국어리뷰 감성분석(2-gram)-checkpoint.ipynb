```python
from konlpy.tag import Twitter
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
import nltk
from nltk.util import ngrams
import pandas as pd
from collections import Counter

review = []
words = []

emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticons
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                u"\U00002702-\U000027B0"
                u"\U0001f926-\U0001f937"
                u'\U00010000-\U0010ffff'
                u"\u200d"
                u"\u2640-\u2642"
                u"\u2600-\u2B55"
                u"\u23cf"
                u"\u23e9"
                u"\u231a"
                u"\u3030"
                u"\ufe0f"
    "]+", flags=re.UNICODE)

twitter= Twitter()

data = pd.read_csv(r"C:\Users\leejisan\Desktop\machine_learning\sample_review\food\SEEN Restaurant & Bar Bangkok.csv",encoding='utf-8',engine='python')
for i in range(len(data)):
    try:
        data['content'][i] = emoji_pattern.sub('', data['content'][i])
    except:
        print(data['content'][i])
        print(i)

stop_words = ['&','것','구','3','2','그',"It's","I'm",'수','5','우리','고','듯','구','거','차','다시','있는','매우','좋습니다'
              ,'의','은','4','편','10','분','개','번','저','여기','다.','30','나',"I've",'수','있습니다','것','중','하나','입니다'
              ,'저희때','라','후','가능','곳','8','만','A5','두','몇','을','한','네','오','1','20','우리','있습니다'
             ,'#1','를','+','aS','25','동안','이곳','시도','때문','이것','일부','때','그것','거리']        
         
text = data['content']

for line in text:
    morph = twitter.pos(line)
    review.append(morph)

for i in review:
    for word,tag in i:
        if tag in ['Noun','Adjective']:
            if not word in stop_words:
                words.append(word)
```

```python
bi_gram = []
bi_tokens = ngrams(words,2)

for i in bi_tokens:
    bi_gram.append(i)
    
sort = sorted(bi_gram, key = lambda x : x[0])
count = Counter(sort)
```

```python
dict = {
    'term1' : [],
    'term2' : [],
    'freq' : []
}

for i,j in count.items():
    dict['term1'].append(list(i)[0])
    dict['term2'].append(list(i)[1])
    dict['freq'].append(j)
    
bi_gram = pd.DataFrame(dict)
bi_gram.sort_values('freq',ascending=False,inplace=True)
bi_gram[0:50]
```

```python
bi_gram['term'] = bi_gram['term1'] + ' ' + bi_gram['term2']
bi_gram[0:50]
```

```python
from wordcloud import WordCloud

bi_gram = bi_gram[['term','freq']]
sample = bi_gram.set_index('term').to_dict()

font_path = "NanumBarunGothic.ttf"  # 한글이 인식 가능한 폰트 설정

wordcloud = WordCloud(
font_path = font_path,
width = 800,
height = 800,
background_color = "white"
)

wordcloud = wordcloud.generate_from_frequencies(sample['freq'])

def to_array(self):
    """Convert to numpy array.
    Returns
    -------
    image : nd-array size (width, height, 3)"
        Word cloud image as numpy matrix.
    """
    return np.array(self.to_image())

array = wordcloud.to_array()

fig = plt.figure(figsize=(10,10))
plt.imshow(array, interpolation="bilinear")  # 보간법 = 쌍선형 보간법
plt.show()
```
