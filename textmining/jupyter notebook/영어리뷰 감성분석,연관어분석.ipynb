```python
import pandas as pd
import re
from afinn import Afinn
from nltk.corpus import stopwords 
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
import numpy as np
import matplotlib.pyplot as plt

sentense = []
count = {}

regex = re.compile("[0-9]")

pos_review=pd.read_csv(r'C:\Users\leejisan\Desktop\sample_review\review\restaurant_review_Khmer Kitchen Restaurant.csv')
pos_review = pos_review[pos_review['lang'] == 'en']
line_pos = pos_review['content']

for i in line_pos:
    sentense.append(i)

sentense = ' '.join(sentense)

tokenizer = RegexpTokenizer('[\w]+')

stop_words = stopwords.words('english')
none_words = ['amok','also','authentic','always']
for i in none_words:
    stop_words.append(i)

for line in line_pos:
    line = line.lower()
    line = regex.sub('',line)
    tokens = tokenizer.tokenize(line)
    stopped_tokens = [i for i in tokens if not i in stop_words]
    for i, a in enumerate(stopped_tokens):
            for b in stopped_tokens[i+1:]:
                if a == b: continue   #같은 단어의 경우는 세지 않음
                if a > b: a, b = b, a   #A, B와 B, A가 다르게 세어지는것을 막기 위해 항상 a < b로 순서 고정
                count[a, b] = count.get((a, b), 0) + 1   #실제로 센다
```

```python
df2=pd.DataFrame(list(count.items()), columns=["term","freq"])

df2=df2.sort_values(by=['freq'],ascending=False)

df2_pos=df2.reset_index(drop=True)
df2_pos.head(50)
```
