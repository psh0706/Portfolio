# 도시별 TF-IDF값 계산 및 Pickle Dump

### City(도시) 컬럼이 있는 전처리 DataFrame import

```python
import pickle

with open("C:\\Users\SOJINSOO\\여행고수\\food_review_merge.txt", "rb") as f:
    food_review_merge= pickle.load(f)
```

### 도시와 식당이 중복된 Review 중복제거

```python
import pandas as pd

food_review_merge.drop_duplicates(['restaurant', 'city'], inplace = True)  # 중복 제거
food_review_merge
```

### 전체 도시 list 확인

```python
food_review_merge['city'].unique()
```

### TF-IDF 계산

```python
# 패키지 import
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import os

# 도시별 TF-IDF 계산
for city in list(food_review_merge['city'].unique()): # 도시 리스트
    
    df_select_region = food_review_merge[food_review_merge['city'].isin([city])] # 해당 도시의 식당들만 추출

    vec = TfidfVectorizer() 
    X = vec.fit_transform(list(df_select_region['content'])) # 해당 도시의 식당들 TF-IDF 계산
    tf_idf = pd.DataFrame(X.toarray(), columns = vec.get_feature_names(), index = list(df_select_region['restaurant']))  # DataFrame 변환
    
    food_review_city = food_review_merge[food_review_merge['city'].isin([city])].reset_index(drop = True) # Index 초기화
    
    try:
        os.mkdir('C:\\Users\\SOJINSOO\\여행고수\\TF-IDF\\' + str(food_review_city['country'][0])) # 해당 디렉토리가 없을 경우 디렉터리 생성
    except:
        pass
    
    print(city)
    
    # 도시별 TF-IDF Pickle Dump 파일 생성 
    with open('C:\\Users\\SOJINSOO\\여행고수\\tf-idf\\'+ 
              str(food_review_city['country'][0]) + 
              '\\' + 
              city + 
              '.txt', 'wb') as f:
        pickle.dump(tf_idf, f)
```
